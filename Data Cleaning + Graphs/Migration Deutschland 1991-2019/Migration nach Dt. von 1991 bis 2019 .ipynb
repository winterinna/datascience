```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
ax=sns.barplot(x = "Jahr", y = "Asylsuchenden", data=data, palette="winter")
ax.set_title("Zahl der Asylsuchenden nach Jahren", fontsize=30)
ax.set_xlabel("Jahre", fontsize=20)
ax.set_ylabel("Anzahl von Asylsuchenden", fontsize=20)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
fig=plt.gcf()
fig.set_size_inches(17,9)
sns.despine(left=True)
```

<p style="font-size:18px">Wie wir sehen können, war die Zahl der Asylsuchenden in Deutschland im Jahr 1992 ungefähr gleich hoch wie im Jahr 2015.<br> 
Im Jahr 2016 im Zuge der Wanderungskrise explodierten die Zahlen, um 2017 wieder rapide zurückzugehen.<br>
Was wir auch erkennen können, ist, dass Kriege verantwortlich sind für Asylzuzug, wie 1991/1992 durch den Jugoslawienkrieg und 2015/2016 durch den Syrienkrieg und weiterer Konflikte in verschiedenen Teilen der Welt.</p>

```python
ax=sns.barplot(x = "Jahr", y = "Zuzug", data=data, palette="winter")
ax.set_title("Zuzug nach Jahren", fontsize=30)
ax.set_xlabel("Jahre", fontsize=20)
ax.set_ylabel("Zuzug in Millionen", fontsize=20)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
fig=plt.gcf()
fig.set_size_inches(17,9)
sns.despine(left=True)
```

<p style="font-size:18px">Das Jahr 1992 sticht heraus: die steigende Zahl ist durch Zuwanderungszahlen durch (Spät-)Aussiedler, Asylbewerber, Flüchtlinge und Arbeitsmigranten von Anfang bis Mitte der 1990er-Jahre zu erklären.<br>
Von 2001 bis 2006 waren die Zuwanderungszahlen insgesamt leicht rückläufig.<br>
Ab 2009 verzeichnen die Zuwanderungszahlen nach Deutschland steigende Tendenz.<br>
Im Jahr 2015 erreicht der Zuzug den höchsten Stand mit 2.016.241 Personen.<br>
    <br>
Interessant ist, dass die Zuzugszahlen ab 2017 stabil bleiben, lediglich mit einer kleiner Verzeichung des Rückgangs im Jahr 2019.<br>
Vergleicht man den Zuzug ab 2017 mit den Asylsuchenden aus diesen Jahren, fällt einem auf, dass der Zuzug nur zu einem kleinen Teil durch Asylbewerber verursacht wird. Über 80% hat andere Gründe für den Zuzug nach Deutschland.</p>

<h2>Gruppe "Zuzug sonstige</h2>
<p style="font-size:18px">Die Gruppe "Zuzug" besteht aus all den Personen, die nach Deutschland gekommen sind. Darunter auch die Asylsuchenden. 
<br>
Interessant ist es zu wissen, wie groß die Gruppe derjenigen ist, die nicht zu den Asylsuchenden gehören. </p>

```python
azz = np.subtract(data["Zuzug"],data["Asylsuchenden"])
azz=pd.DataFrame(azz)
azz.columns=['Sonstige']
azz.head()
```

```python
data1=pd.concat([data,azz], axis=1)
data1.head()
```

```python
ax=sns.barplot(x = "Jahr", y = "Sonstige", data=data1, palette="winter")
ax.set_title("Zuzug Sonstige", fontsize=30)
ax.set_xlabel("Jahre", fontsize=20)
ax.set_ylabel("Zuzug in Millionen", fontsize=20)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
fig=plt.gcf()
fig.set_size_inches(17,9)
sns.despine(left=True)
```

<p style="font-size:18px">Ab dem Jahr 2010 nimmt die Anzahl der Gruppe "Zuzug sonstige" zu. </p>
